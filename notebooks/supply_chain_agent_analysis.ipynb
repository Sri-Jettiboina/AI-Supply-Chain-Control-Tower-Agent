# AI Supply Chain Control Tower Agent Analysis

Use this notebook to explore supplier risk, stockout risk, and procurement spend before building the Streamlit app.


```python
import pandas as pd
df = pd.read_csv('../data/cleaned_supply_chain_data.csv')
df.head()
```

```python
df.groupby('supplier_name').agg(orders=('po_id','count'), spend=('po_value','sum')).sort_values('spend', ascending=False).head(10)
```

```python
late_rate = (df['delivery_status'].eq('Late').mean()) * 100
defect_rate = (df['defective_units'].sum() / df['quantity_received'].sum()) * 100
late_rate, defect_rate
```
